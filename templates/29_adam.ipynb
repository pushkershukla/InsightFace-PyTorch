````python
# вњЏпёЏ YOUR IMPLEMENTATION HERE

class MyAdam:
    def __init__(self, params, lr=1e-3, betas=(0.9, 0.999), eps=1e-8):
        self.params=params
        self.lr=lr
        self.beta_1=betas[0]
        self.beta_2=betas[1]
        self.eps=eps
        self.m=torch.zeros_like(params)
        self.v=torch.zeros_like(params)
        pass  # store params, init m and v to zeros

    def step(self):
        self.m=self.beta_1*self.m + (1-self.beta_1)*self.params.gradients
        self.v=self.beta_2*self.v +(1-self.beta_2)*self.params.gradients
        self.m=self.m/(1-self.beta_1+self.eps)
        self.v=self.v/(1-self.beta_2+self.eps)
        self.params-=self.lr*self.m/(torch.sqrt(self.v)+self.eps)
        pass  # update params using Adam rule

    def zero_grad(self):
        self.params.gradients=torch.zeros_like(self.params.gradients)
        pass  # zero all gradients
````

````python
# рџ§Є Debug
torch.manual_seed(0)
w = torch.randn(4, 3, requires_grad=True)
opt = MyAdam([w], lr=0.01)
for i in range(5):
    loss = (w ** 2).sum()
    loss.backward()
    opt.step()
    opt.zero_grad()
    print(f'Step {i}: loss={loss.item():.4f}')
````
